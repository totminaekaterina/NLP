```python
import pandas as pd
import numpy as np

# Заданные данные для последней точки
last_data = {
    "DEPT (м)": 440.0,
    "GKIN (мкР/час)": 82.7,
    "K (имп/мин)": 986,
    "TH (имп/мин)": 1114,
    "U (имп/мин)": 1177,
    "KMV (ед.СИ*10^-5)": 98.6,
    "ИК (мСим/м)": 96.3,
    "CM (нТл)": 5029
}

# Создаем DataFrame
data = pd.DataFrame(last_data, index=[0])

# Генерация данных для следующих 10 точек (440.5 м, 441.0 м, ... до 445.0 м)
for depth in np.arange(440.5, 445.5, 0.5):
    new_data = {}
    new_data["DEPT (м)"] = depth
    for key in last_data:
        if key != "DEPT (м)":
            # Добавляем случайные колебания в пределах +- 0.5% от последнего значения
            variation = np.random.uniform(-0.005, 0.005) * last_data[key]
            new_data[key] = last_data[key] + variation
            # Обновляем последнее значение
            last_data[key] = new_data[key]
    data = data.append(new_data, ignore_index=True)

data
```

```python
# Измененный код для генерации данных с одной цифрой после запятой

import pandas as pd
import numpy as np

# Предполагаем, что переменная data уже содержит некоторые сгенерированные данные
# Пример исходных данных
data_example = {
    "DEPT (м)": [436.0, 436.5, 437.0, 437.5, 438.0, 438.5, 439.0, 439.5, 440.0],
    "GKIN (мкР/час)": [81.9, 82.0, 82.1, 82.2, 82.3, 82.4, 82.5, 82.6, 82.7],
    "K (имп/мин)": [974, 975, 977, 978, 980, 981, 983, 984, 986],
    "TH (имп/мин)": [1106, 1107, 1108, 1108, 1109, 1110, 1111, 1112, 1113],
    "U (имп/мин)": [1165, 1166, 1168, 1169, 1171, 1172, 1174, 1175, 1177],
    "KMV (ед.СИ*10^-5)": [97.5, 97.6, 97.8, 97.9, 98.0, 98.2, 98.3, 98.4, 98.6],
    "ИК (мСим/м)": [95.1, 95.3, 95.4, 95.6, 95.7, 95.9, 96.0, 96.2, 96.3],
    "CM (нТл)": [4965, 4973, 4981, 4989, 4997, 5005, 5013, 5021, 5029]
}

# Создаем DataFrame из исходных данных
data = pd.DataFrame(data_example)

# Обновляем последние известные значения
last_data = data.iloc[-1].to_dict()

# Инициализация пустого списка для хранения новых данных
new_data_list = []

# Генерация данных для следующих 10 точек (445.5 м, 446.0 м, ... до 450.0 м)
for depth in np.arange(440.5, 6000., 0.5):
    new_data = {"DEPT (м)": round(depth, 1)}
    for key in last_data:
        if key != "DEPT (м)":
            # Добавляем случайные колебания в пределах +- 0.5% от последнего значения
            variation = np.random.uniform(-0.005, 0.005) * last_data[key]
            new_value = last_data[key] + variation
            new_data[key] = round(new_value, 1)  # Округление до одной цифры после запятой
            # Обновляем последнее значение
            last_data[key] = new_value
    new_data_list.append(new_data)

# Конвертация списка новых данных в DataFrame и объединение с исходным DataFrame
new_data_df = pd.DataFrame(new_data_list)
data = pd.concat([data, new_data_df], ignore_index=True)

# Показываем сгенерированные данные
data.tail(5)  # Показываем последние 10 строк данных
```

```python
data.to_csv("pr_1_data.csv", index=False, encoding='utf-8', sep=';')
```
